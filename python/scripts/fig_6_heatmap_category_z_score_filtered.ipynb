```python
%pip install pandas numpy seaborn matplotlib --quiet

import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# --- Folder with CSV files ---

input_dir = "/Users/jeanclaudemakangara/Library/CloudStorage/OneDrive-UniversitaetBern/ISPM_PhD_Project_Makangara_Cigolo_Jean_Claude/Writing_manuscript/Protocol_comparison_Panzi/protocol_comparison_Panzi_manuscript_Python_script/data"
output_dir = "/Users/jeanclaudemakangara/Library/CloudStorage/OneDrive-UniversitaetBern/ISPM_PhD_Project_Makangara_Cigolo_Jean_Claude/Writing_manuscript/Protocol_comparison_Panzi/protocol_comparison_Panzi_manuscript_Python_script/outputs"
os.makedirs(output_dir, exist_ok=True)

csv_files = [f for f in os.listdir(input_dir) if f.endswith('.csv')]

selected_names = ["virus", "bacteria", "eukaryota", "archaea"]
protocol_order = ['eSISPA', 'PtCl4 eSISPA', 'Standard SISPA']

for file in csv_files:
    print(f"\nProcessing: {file}")
    df = pd.read_csv(os.path.join(input_dir, file))
    df.columns = [col.strip().lower() for col in df.columns]
    df['protocol'] = df['protocol'].str.strip()
    df['sample_type'] = df['sample_type'].str.strip()
    
    df_selected = df[df['category'].isin(selected_names)]
    df_selected = df_selected[np.log10(df_selected['z_score'] + 1) > 0.29]

    for protocol in protocol_order:
        samples_in_protocol = df[df['protocol'] == protocol]['short_name_id'].unique()
        df_protocol = df_selected[df_selected['protocol'] == protocol]

        # Define sample_type safely before plotting
        if not df_protocol.empty and 'sample_type' in df_protocol.columns:
            sample_type = df_protocol['sample_type'].iloc[0]
        else:
            sample_type = "Unknown"

        heat_df = df_protocol.pivot_table(
            index='category',
            columns='short_name_id',
            values='nt_bpm',
            aggfunc='sum'
        ).reindex(index=selected_names, columns=samples_in_protocol)

        sorted_samples = sorted(samples_in_protocol)
        heat_df = heat_df.reindex(columns=sorted_samples)

        log_heat_df = np.log10(heat_df + 1)
        missing_mask = heat_df.isna()
        display_df = log_heat_df.fillna(0).round(2)

        print(f"\n--- Protocol: {protocol} ---")
        print(display_df)

        plt.figure(figsize=(max(10, len(sorted_samples) * 0.6), 6))
        ax = sns.heatmap(
            display_df,
            cmap='YlGn',
            linewidths=1,
            linecolor='grey',
            cbar=True,
            annot=True,
            fmt='.2f',
            vmin=0,
            vmax=6,
            xticklabels=True,
            yticklabels=True,
            mask=missing_mask
        )

        cbar = ax.collections[0].colorbar
        cbar.set_label('log10(nt_bpm + 1)', fontsize=13, fontweight='bold')

        plt.xticks(rotation=90, ha='right', fontsize=11)
        plt.yticks(rotation=0, fontsize=12)
        plt.xlabel("Sample", fontsize=13, fontweight='bold')
        plt.ylabel("")
        plt.title(
            f"Taxa categories heatmap (raw data with z-score > 1) for {sample_type}\nProtocol: {protocol}",
            fontsize=15, fontweight='bold'
        )

        plt.tight_layout()
        # Save figure to output directory, prefixing with 'fig_6_heat_filtered' and ensure unique names
        base_name = f"fig_6_heatmap_filtered_{protocol.replace(' ', '_')}"
        outname_base = base_name
        # If a file with the same name already exists, append a letter suffix (_a, _b, ...)
        for letter in [""] + [chr(i) for i in range(ord('a'), ord('z')+1)]:
            candidate = outname_base + (f"_{letter}" if letter else "")
            png_path = os.path.join(output_dir, candidate + ".png")
            pdf_path = os.path.join(output_dir, candidate + ".pdf")
            if not (os.path.exists(png_path) or os.path.exists(pdf_path)):
                outname_base = candidate
                break

        plt.savefig(os.path.join(output_dir, outname_base + ".png"), dpi=600, bbox_inches='tight')
        plt.savefig(os.path.join(output_dir, outname_base + ".pdf"), bbox_inches='tight')
        plt.show()
        plt.close()
```
