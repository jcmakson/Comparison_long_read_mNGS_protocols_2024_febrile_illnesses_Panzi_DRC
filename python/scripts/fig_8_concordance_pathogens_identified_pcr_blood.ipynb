```python
%pip install pandas numpy seaborn matplotlib --quiet

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
import matplotlib.gridspec as gridspec

# --- Set input and output directories ---
inputdir = Path("/Users/jeanclaudemakangara/Library/CloudStorage/OneDrive-UniversitaetBern/ISPM_PhD_Project_Makangara_Cigolo_Jean_Claude/Writing_manuscript/Protocol_comparison_Panzi/protocol_comparison_Panzi_manuscript_Python_script/data")
outputdir = Path("/Users/jeanclaudemakangara/Library/CloudStorage/OneDrive-UniversitaetBern/ISPM_PhD_Project_Makangara_Cigolo_Jean_Claude/Writing_manuscript/Protocol_comparison_Panzi/protocol_comparison_Panzi_manuscript_Python_script/outputs")
outputdir.mkdir(parents=True, exist_ok=True)

# --- Load your data ---
df = pd.read_csv(inputdir / "fig_7_8_z_blood_samples.csv")

# --- Clean and standardize columns ---
df.columns = [col.strip().lower() for col in df.columns]
df['protocol'] = df['protocol'].str.strip()

# --- Define PCR-confirmed pathogens of interest ---
selected_names = ["Plasmodium", "Salmonella", "Klebsiella", "Pseudomonas"]

# --- Keep only rows for selected taxa ---
df_selected = df[df['name'].isin(selected_names)]

# --- Apply minimum abundance filter ---
#df_selected = df_selected[df_selected['nt_bpm'] >= 100]

# --- Apply z-score > 1 filter ---
if 'z_score' in df_selected.columns:
    df_selected = df_selected[df_selected['z_score'] > 1]
else:
    print("Warning: 'z_score' column not found. No z-score filtering applied.")

# --- List of protocols to compare ---
protocol_order = ['eSISPA', 'PtCl4 eSISPA', 'Standard SISPA']

# --- Generate heatmaps for all protocols in a single figure ---
fig = plt.figure(figsize=(max(10, len(protocol_order) * 8), 7))
gs = gridspec.GridSpec(1, len(protocol_order), wspace=0.3)
axes = []

for i, protocol in enumerate(protocol_order):
    samples_in_protocol = df[df['protocol'] == protocol]['short_name_id'].unique()
    df_protocol = df_selected[df_selected['protocol'] == protocol]
    heat_df = df_protocol.pivot_table(
        index='name',
        columns='short_name_id',
        values='nt_bpm',
        aggfunc='sum'
    ).reindex(index=selected_names, columns=samples_in_protocol)
    sorted_samples = sorted(samples_in_protocol)
    heat_df = heat_df.reindex(columns=sorted_samples)
    log_heat_df = np.log10(heat_df + 1)
    missing_mask = heat_df.isna()
    display_df = log_heat_df.fillna(0).round(2)

    ax = fig.add_subplot(gs[0, i])
    axes.append(ax)
    sns.heatmap(
        display_df,
        cmap='YlGn',
        linewidths=1,
        linecolor='grey',
        cbar=(i == len(protocol_order)-1),  # Only show colorbar on last subplot
        annot=True,
        fmt='.2f',
        vmin=0,
        vmax=6,
        xticklabels=True,
        yticklabels=True,
        mask=missing_mask,
        ax=ax
    )
    ax.set_title(f"{protocol}", fontsize=14, fontweight='bold')
    ax.set_xlabel("Sample", fontsize=12, fontweight='bold')
    if i == 0:
        ax.set_ylabel("Pathogen", fontsize=12, fontweight='bold')
    else:
        ax.set_ylabel("")
    ax.tick_params(axis='x', rotation=90, labelsize=10)
    ax.tick_params(axis='y', rotation=0, labelsize=11)

# Set main title
fig.suptitle("Selected pathogens in blood samples (z-score > 1)", fontsize=16, fontweight='bold', y=1.05)
plt.tight_layout(rect=[0, 0, 1, 0.97])

# Add colorbar label if present
if axes:
    cbar = axes[-1].collections[0].colorbar
    cbar.set_label('log10(nt_bpm + 1)', fontsize=13, fontweight='bold')

# Save combined figure
fig.savefig(outputdir / "fig_8_concordance_pcr_blood.png", dpi=600, bbox_inches='tight')
fig.savefig(outputdir / "fig_8_concordance_pcr_blood.pdf", bbox_inches='tight')
plt.show()
plt.close(fig)
```
